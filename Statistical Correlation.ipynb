```python
import pandas as pd
import numpy as np
```

```python
dat = pd.read_csv("/home/donghan/Downloads/C.csv", header = None)
```

```python
df = dat.T
df.columns = list(range(1, len(dat.index)+1))
```

```python
corr = df.corr()
```

```python
for i in range(1, len(corr.columns)+1):
    boolean = list(corr[i].apply(lambda x: True if x >= 0.75 else False))
    for j in range(len(boolean)):
        if boolean[j] == True and corr[i][j+1] != 1.0:
            print(i, j+1, corr[i][j+1])
```

```python
import cv2
import os

os.chdir("/home/donghan/DeepLabCut/data/")
#Change to your own video data directory

def videoWriterGray(task, filename, outputName):
    
    '''
    Task: Takes video as input and returns its capturing and output file. 
    
    If task is edited, then corrX and corrY is the vertex of the cropped video edge. 
    '''

    # Capture video
    cap = cv2.VideoCapture(filename)

    # Read video frame by frame
    # Extract original video frame features
    sz = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
            int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

    fourcc = int(cap.get(cv2.CAP_PROP_FOURCC))

    fps = int(cap.get(cv2.CAP_PROP_FPS))
    # Make a directory to store the processed videos
    path = "./" + task
    try:  
        os.mkdir(path)
        print ("Successfully created the directory %s " % path)
    except OSError:  
        pass
    
    #Automatically name the processed videos  
    file = "./" + task + "/" + outputName

    out = cv2.VideoWriter(file, cv2.VideoWriter_fourcc(*'XVID'), fps, sz)
    # Another option MJPG

    return cap,out
    
    
    
def color2Gray(task, filename, outputName, displayVideo = False):
    
    '''
    Task: Takes video as input and returns its greyscale version. 
    '''

    video = videoWriterGray(task, filename, outputName)
    cap = video[0]
    out = video[1]
    frameCnt = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    # Not capturing a video
    if (cap.isOpened() == False): 
          print("Unable to read video")
    count = 1
    while(cap.isOpened()):
        ret,img = cap.read()
        try:
            # Convert to grayscale
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            # In order to save as 3-channel video, ex, RGB, convert back to RGB, but remains gray. 
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
            out.write(img)
            if displayVideo == True:
                cv2.imshow('edited video',img) 
            if count == frameCnt:
                print (filename, 'successfully ', "edited!")
                break
            else:
                count += 1

            k=cv2.waitKey(30) & 0xff
            #once you inter Esc capturing will stop
            if k==27:
                break
        except:
            break
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    
    
    

color2Gray("color2Gray", '1034 SI_B, Aug 15, 13 7 49.mp4', "color2GrayTest.mp4", False)
```
